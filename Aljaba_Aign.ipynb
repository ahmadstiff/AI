```python
# Langkah 1: Import pustaka yang diperlukan
#22.11.4784 Ahmad Aroyanu Staifen
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
from google.colab import files
import matplotlib.pyplot as plt

# Langkah 2: Muat model VGG16 dengan bobot yang telah dilatih pada dataset ImageNet
model = VGG16(weights='imagenet')
```

```python
uploaded = files.upload()
```

```python
image_path = list(uploaded.keys())[0]
```

```python
img = image.load_img(image_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
predictions = model.predict(x)
predicted_classes = decode_predictions(predictions, top=3)[0]
```

```python
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python

for i, (imagenet_id, label, score) in enumerate(predicted_classes):
    print(f"{i + 1}: {label} ({score * 100:.2f}%)")
```
